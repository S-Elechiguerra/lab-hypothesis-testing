- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
# Dragon + Grass

import pandas as pd
import scipy.stats as stats

import pandas as pd
import scipy.stats as stats

# Dragon-type and Grass-type Pokémon
dragon_hp = df[df['Type 2'] == 'Dragon']['HP']
grass_hp = df[df['Type 1'] == 'Grass']['HP']

# Conduct the Two Sample T-Test
t_stat, p_value = stats.ttest_ind(dragon_hp, grass_hp, equal_var=True)

# Since we are testing if Dragon HP is greater than Grass HP, we need a one-tailed test
p_value_one_tailed = p_value / 2 if t_stat > 0 else 1 - (p_value / 2)

# Print results
print(f"T-statistic: {t_stat}")
print(f"P-value (one-tailed): {p_value_one_tailed}")

# Decision
if p_value_one_tailed < 0.05:
    print("Reject the null hypothesis: Dragon-type Pokémon have significantly higher HP.")
else:
    print("Fail to reject the null hypothesis: No significant difference in HP.")


```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
# ANOVA 

import pandas as pd
from scipy.stats import f_oneway  

# List of stats to analyze
stats_columns = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

# Perform ANOVA for each stat
for stat in stats_columns:
    legendary_stats = df[df['Legendary'] == 'True'][stat]
    non_legendary_stats = df[df['Legendary'] == 'False'][stat]

    # Perform one-way ANOVA
    f_stat, p_value = f_oneway(legendary_stats, non_legendary_stats)
    
    print(f"ANOVA test for {stat}: F-statistic = {f_stat}, p-value = {p_value}")

    if p_value < 0.05:
        print(f"Reject the null hypothesis for {stat}: Significant difference exists.\n")
    else:
        print(f"Fail to reject the null hypothesis for {stat}: No significant difference.\n")
```

```python
from scipy.stats import kruskal

# Conduct the Kruskal-Wallis test for each stat
for stat in ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']:
    legendary_stats = df[df['Legendary'] == 'True'][stat]
    non_legendary_stats = df[df['Legendary'] == 'False'][stat]

    h_stat, p_value = kruskal(legendary_stats, non_legendary_stats)
    
    print(f"Kruskal-Wallis test for {stat}: H-statistic = {h_stat}, p-value = {p_value}")

    if p_value < 0.05:
        print(f"Reject the null hypothesis for {stat}: Significant difference exists.\n")
    else:
        print(f"Fail to reject the null hypothesis for {stat}: No significant difference.\n")
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 37)
- Hospital coordinates (-122, 34)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
import pandas as pd
import numpy as np
from scipy import stats

# Coordinates for school and hospital
school_coords = (-118, 37)
hospital_coords = (-122, 34)

# Function to calculate Euclidean distance
def calculate_distance(row, coords):
    return np.sqrt((row[0] - coords[0])**2 + (row[1] - coords[1])**2)

# Calculate distances to school and hospital
df['distance_to_school'] = df.apply(calculate_distance, axis=1, args=(school_coords,))
df['distance_to_hospital'] = df.apply(calculate_distance, axis=1, args=(hospital_coords,))

# Define threshold distance
threshold_distance = 0.50

# Create a new column to identify close houses
df['close_to_school'] = df['distance_to_school'] < threshold_distance
df['close_to_hospital'] = df['distance_to_hospital'] < threshold_distance
df['close_to_either'] = df['close_to_school'] | df['close_to_hospital']

# Prices of the two groups
close_prices = df[df['close_to_either']]['median_house_value']
far_prices = df[~df['close_to_either']]['median_house_value']

# Perform independent t-test
t_stat, p_value = stats.ttest_ind(close_prices, far_prices, equal_var=False)

# Output results
alpha = 0.05  # Significance level
print(f"T-statistic: {t_stat}, P-value: {p_value}")

# Conclusion
if p_value < alpha:
    conclusion = "Reject the null hypothesis: There is a significant difference in prices."
else:
    conclusion = "Fail to reject the null hypothesis: There is no significant difference in prices."

print(conclusion)
```
